```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.normal(170, 10, 1000)
```

```python
plt.figure(figsize = (12,8))

ax = sns.histplot(data, kde="true")

line = ax.get_lines()[-1]
x, y = line.get_data()

mask = (x >= 180) & (x <= 190)
x, y = x[mask], y[mask]

ax.fill_between(x, y1=y, alpha=0.5, facecolor='red')

plt.show() 
```

```python
plt.figure(figsize = (12,8))
hist = plt.hist(data, bins=30)
```

```python
density, bins, patches = hist
widths = bins[1:] - bins[:-1]
total_area = (density * widths).sum()
```

```python
sub_area = 0

for bin, density in list(zip(bins, density)):
    if (bin >= 180) and (bin <= 190):
        sub_area += density * widths[0]
```

```python
sub_area/total_area
```
